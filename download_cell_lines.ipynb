```python
def QuerySetCellLinesFromIdList(cell_line_chembl_id_list: list[str]) -> QuerySet:
    return new_client.cell_line.filter(cell_chembl_id__in=cell_line_chembl_id_list) # type: ignore
```

```python
cell_line_chembl_id_list: list[str] = ["CHEMBL4295386", "CHEMBL3307781", "CHEMBL4295453", "CHEMBL4295483", "CHEMBL3308509", "CHEMBL3706569", "CHEMBL3307715", 
                  "CHEMBL3307525", "CHEMBL3307970", "CHEMBL4295409", "CHEMBL3307501", "CHEMBL3307364", "CHEMBL3308499", "CHEMBL3307481", "CHEMBL3308021", "CHEMBL3307755", "CHEMBL3307614"]
```

```python
df = pd.DataFrame(QuerySetCellLinesFromIdList(cell_line_chembl_id_list)) # type: ignore

df
```

```python
def LinkToActivitiesRelatedToCellLine(cell_line_id: str, relation_type: str) -> str:
  return f"https://www.ebi.ac.uk/chembl/web_components/explore/activities/QUERYSTRING:_metadata.assay_data.cell_chembl_id%3A{cell_line_id}%20AND%20standard_type%3A%22{relation_type}%22?description=Bioactivity%20data%20for%20Cell%20Line%20{cell_line_id}%20-%20{relation_type}"
```

```python
cell_line_activities_links: dict[str, dict[str, str]] = {}

for cell_line_chembl_id in cell_line_chembl_id_list:
    cell_line_activities_links[cell_line_chembl_id] = {
        "IC50": LinkToActivitiesRelatedToCellLine(cell_line_chembl_id, "IC50"),
        "GI50": LinkToActivitiesRelatedToCellLine(cell_line_chembl_id, "GI50"),
}
```

```python
for cell_line_chembl_id, links in cell_line_activities_links.items():
    print(f"Cell line: {cell_line_chembl_id}")
    print(f"IC50: {links['IC50']}")
    print(f"GI50: {links['GI50']}")
    print("-" * 20)
```
